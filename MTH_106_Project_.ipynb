```python
def cosineSimilarity(u, v):
  return np.dot(u, v) / ( norm(u) * norm(v) )
```

```python
df_ForthM = pd.DataFrame({
    'Tot sales this month (4) Generated': np.random.randint(1, 1800, 36),
})
```

```python
df_ForthM
```

```python
arr_ForthM = df_ForthM.to_numpy()
```

```python
 a = cosineSimilarity(arr_firstM , arr_ForthM)
```

```python
 b = cosineSimilarity(arr_secondM , arr_ForthM)
```

```python

 c= cosineSimilarity(arr_thirdM , arr_ForthM)
```

```python
MaxSim = -9999
CosSim = a

if CosSim > MaxSim:
    MaxSim = CosSim
    pos = 'Tot sales this month (1)'
    print("Maximum similarity:",pos , MaxSim)
```

```python
CosSim = b

if CosSim > MaxSim:
    pos = 'Tot sales this month (2)'
    MaxSim = CosSim
    print("Maximum similarity:",pos, MaxSim)
```

```python
CosSim = c

if CosSim > MaxSim:
    pos = 'Tot sales this month (3)'
    MaxSim = CosSim
    print("Maximum similarity:",pos ,MaxSim)
```
